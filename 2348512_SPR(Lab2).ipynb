**(1) Question 1 :**

```python
# (a) Generate a basic sinusoidal signal in the time domain:
# Let's generate a sine wave signal with a frequency of 5Hz, sampled at 1000Hz. This means that for every second, we will have 1000 samples. The equation for a sine wave is: y(t) = A*sin(2Ï€ft), where A is the amplitude, f is the frequency in Hz, and t is time.

import numpy as np
import matplotlib.pyplot as plt

# Define parameters
amplitude = 1
frequency = 5  # Hz
sampling_rate = 1000  # Hz
num_samples = 1000

# Generate the time vector
time = np.arange(num_samples) / sampling_rate

# Generate the sinusoidal signal
signal = amplitude * np.sin(2 * np.pi * frequency * time)
```

```python
# (b) Plot the time-domain waveform of the signal:

plt.figure(figsize=(12, 6))
plt.plot(time, signal)
plt.title('Time Domain Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum:
#The Discrete-Time Fourier Transform (DTFT) is a continuous transformation of a discrete and periodic function. In practice, we can approximate the DTFT using the Fast Fourier Transform (FFT).

dtft = np.fft.fft(signal)

# Frequency axis for DTFT
freq_dtft = np.fft.fftfreq(num_samples, d=1/sampling_rate)

plt.figure(figsize=(12, 6))
plt.plot(freq_dtft, np.abs(dtft))
plt.title('DTFT (Continuous Frequency Spectrum)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
# (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum:
#The Discrete Fourier Transform (DFT) is a sampled version of the DTFT. We can directly compute the DFT using the FFT function.


dft = np.fft.fft(signal)
freq_dft = np.fft.fftfreq(num_samples, d=1/sampling_rate)

plt.figure(figsize=(12, 6))
plt.stem(freq_dft, np.abs(dft), use_line_collection=True)
plt.title('DFT (Discrete Frequency Spectrum)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

**(2) Question 2 :**

```python
# a) Generate a composite signal composed of two sinusoidal signals: Let's generate a composite signal by adding two sinusoidal signals with different frequencies and amplitudes. You can adjust the frequencies and amplitudes as needed.

# Define parameters for the sinusoidal signals
amplitude_1 = 1.5
frequency_1 = 10  # Hz
amplitude_2 = 0.7
frequency_2 = 3  # Hz

# Generate the two sinusoidal signals
signal_1 = amplitude_1 * np.sin(2 * np.pi * frequency_1 * time)
signal_2 = amplitude_2 * np.sin(2 * np.pi * frequency_2 * time)

# Create the composite signal by adding the two sinusoidal signals
composite_signal = signal_1 + signal_2
```

```python
# (b) Plot the time-domain waveform of the composite signal: Let's plot the composite signal in the time domain:

plt.figure(figsize=(12, 6))
plt.plot(time, signal)
plt.title('Time Domain Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum: Compute the DTFT of the composite signal and plot the continuous frequency spectrum:

dtft_comp = np.fft.fft(composite_signal)
freq_dtft_comp = np.fft.fftfreq(num_samples, d=1/sampling_rate)

plt.figure(figsize=(12, 6))
plt.plot(freq_dtft_comp, np.abs(dtft_comp))
plt.title('DTFT of Composite Signal (Continuous Frequency Spectrum)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
# (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum: Compute the DFT of the composite signal and plot the discrete frequency spectrum:

dft_comp = np.fft.fft(composite_signal)
freq_dft_comp = np.fft.fftfreq(num_samples, d=1/sampling_rate)

plt.figure(figsize=(12, 6))
plt.stem(freq_dft_comp, np.abs(dft_comp), use_line_collection=True)
plt.title('DFT of Composite Signal (Discrete Frequency Spectrum)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Fs = 1000  # Sampling frequency
T = 1      # Duration of signal in seconds
t = np.linspace(0, T, T*Fs, endpoint=False)  # Time vector
a = 2      # Decay constant
```

```python
# (a) Generate an exponentially decaying signal
x = np.exp(-a*t)
```

```python
# (b) Plot the time-domain waveform
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Exponentially Decaying Signal in Time Domain')
```

```python
# (c) Compute the Discrete-Time Fourier Transform (DTFT) and plot the continuous frequency spectrum
w = np.linspace(-np.pi, np.pi, len(t))
X_DTFT = np.fft.fftshift(np.fft.fft(x))
plt.subplot(3, 1, 2)
plt.plot(w, abs(X_DTFT))
plt.xlabel('Frequency (radians/sample)')
plt.ylabel('Magnitude')
plt.title('Continuous Frequency Spectrum (DTFT)')
```

```python
# (d) Compute the Discrete Fourier Transform (DFT) and plot the discrete frequency spectrum
X_DFT = np.fft.fft(x)
f = np.arange(0, Fs, Fs/len(t))
plt.subplot(3, 1, 3)
plt.stem(f, abs(X_DFT))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('Discrete Frequency Spectrum (DFT)')

plt.tight_layout()
plt.show()
```
